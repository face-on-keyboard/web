```python
import pandas as pd
```

```python
df = pd.read_csv('carbon-label.csv', header=[0, 1])

df.columns = [col[1] if col[1][0] != 'U' else col[0] for col in df.columns]

df.columns = [col.rstrip('_') for col in df.columns]

df.head()
```

```python
df.columns
```

```python
df.columns = [
  "productType",
  "productName",
  "productModel",
  "證書編號",
  "生效日期",
  "有效期限",
  "status",
  "companyName",
  "uniformNumber",
  "名稱1",
  "地址1",
  "名稱2",
  "地址2",
  "名稱3",
  "地址3",
  "carbonFootprint",
  "declarationUnit",
  '原料取得', 
  '製造/服務', 
  '配送銷售', 
  '使用', 
  '廢棄回收', 
  '編號', 
  '名稱'
]
```

```python
df.head()
```

```python
df.dropna(subset=["productName", "productType", "carbonFootprint"], inplace=True)
```

```python
df.to_json('carbon-label.json', force_ascii=False, orient='records')
```
